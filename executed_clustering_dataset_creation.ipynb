# Clustering Analysis Dataset Creation

This notebook demonstrates how to create a synthetic dataset suitable for clustering analysis using `sklearn.datasets.make_blobs`. It also includes visualization of the generated clusters and saving the dataset to a CSV file.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset with 3 clusters and 300 samples
X, y = make_blobs(n_samples=300, centers=3, n_features=2, cluster_std=1.0, random_state=42)

# Convert to DataFrame for easier handling
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
df['Cluster'] = y

# Display first few rows
df.head()
```

```python
# Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(df['Feature_1'], df['Feature_2'], c=df['Cluster'], cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('Synthetic Clustering Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Save the dataset to a CSV file
df.to_csv('clustering_dataset.csv', index=False)
print('Dataset saved to clustering_dataset.csv')
```
